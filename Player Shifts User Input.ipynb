```python
# Import packages

import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pybaseball as pyb
from pybaseball import playerid_lookup
```

```python
# Collect all data from the past 3 years

data = pyb.statcast(start_dt = '2015-04-01', end_dt = '2022-09-30')
```

```python
# Only look at balls hit into play

bip = data[data.description == 'hit_into_play']
```

```python
# Replace all strings with integer values for easier data analysis 

bip = bip.replace('single', 1)
bip = bip.replace('double', 2)
bip = bip.replace('triple', 3)
bip = bip.replace('home_run', 4)
bip = bip.replace('field_out', -1)
bip = bip.replace('force_out', -1)
bip = bip.replace('fielders_choice_out', -1)
bip = bip.replace('sac_bunt', -1)
bip = bip.replace('grounded_into_double_play', -2)
bip = bip.replace('double_play', -2)
bip = bip.replace('sac_fly', -1)
bip = bip.replace('field_error', 0)
bip = bip.replace('catcher_interf', 0)
bip = bip.replace('fielders_choice', -1)
bip = bip.replace('sac_fly_double_play', -2)
bip = bip.replace('sac_bunt_double_play', -2)
bip = bip.replace('triple_play', -3)
```

```python
#interpolated
bip['hc_x1'] = bip['hc_x'].round()
bip['hc_y1'] = bip['hc_y'].round()
```

```python
def create_pd_player(first_name, last_name):
    """ Returns a dataframe for a given player
    Args:
        player_number(int): a player_id
    
    Returns:
        pd_player(df): dataframe with all appropriate data for a player
    """
    # Create a new dataframe for the player
    pd_player = pd.DataFrame(columns = bip.columns)
    
    # Find all instances of balls hit by the player
    pd_player = bip[bip['batter'] == int(playerid_lookup(last_name,first_name )['key_mlbam'])]


    # Return df
    return pd_player
```

```python
def create_visual(df, x, y , z,aggfunc = 'mean'):
    """ Creates a heatmap of the given data
    
    Args:
        df(df): dataframe
        x(str): x axis
        y(str): y axis
        z(str): z axis
        aggfunc(str): aggregate function to find for pivoting
    
    Returns:
        ax(heatmap): heatmap
    """
    bip_2 = df[[x,y,z]]
    bip_2 = bip_2.dropna()
    bip_2 = bip_2.pivot_table(index=x, columns=y, values=z, aggfunc=aggfunc)
    ax = sns.heatmap(bip_2)
```

```python
# User inputs first and last name
first_name = input("Enter First Name: ")
last_name = input("Enter Last Name: ")

# Find all data correlating to inputted player
pd_player = create_pd_player(first_name, last_name)

# Create a heatmap for the chosen player
create_visual(pd_player, 'hc_x1', 'hc_y1', 'events')
```
